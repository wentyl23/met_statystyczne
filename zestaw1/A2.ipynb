```python
import random
import matplotlib.pyplot as plt
import math
```

### Problem A2

```python
starting_points=[0,0,-2]
scales=[0.5,1.0,1.0]

def density_func(y,y_zero,scale):
    return 1/(math.pi*scale*(1+((y-y_zero)/scale)**2))
    
def inversed_distribution(x,x_zero,scale):
    return scale*math.tan(math.pi*x-1/2*math.pi) + x_zero

def average(x,N):
    return sum(x)/N
def variation(x,avg,N):
    result=0
    for i in range(N):
        result+=(x[i]-avg)**2
    return result/N

N=10000
xs=[]
ys=[]
args=[]

for start,scale in zip(starting_points,scales):
    arg=[]
    y=[]
    x=[]
    for j in range(N):
        arg.append(-5+j*0.001)
        y.append(inversed_distribution(random.random(),start,scale))
        x.append(density_func(arg[j],start,scale))
    args.append(arg)
    ys.append(y)
    xs.append(x)
fig,a =  plt.subplots(len(scales),figsize=(20,20))

for i in range(len(scales)):
    a[i].hist(ys[i],range=[-5, 5], density=True, bins=30)
    a[i].plot(args[i],xs[i],color = 'red')
    print("wariancja: " + str(variation(xs[i],average(xs[i],N),N)) + " srednia: " + str(average(xs[i],N)))

```
