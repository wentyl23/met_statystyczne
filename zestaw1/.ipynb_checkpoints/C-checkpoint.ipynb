```python
import random
import matplotlib.pyplot as plt
import math
```

### Problem B

```python
def simulate_game(a,b,p,N):
    q=1-p
    a_ruins=0
    b_ruins=0
    for i in range(N):
        curr_a = a
        curr_b = b
        while curr_a > 0 and curr_b > 0:
            if random.random() < p :
                curr_a+=1
                curr_b-=1
            else:
                curr_a-=1
                curr_b+=1
        if curr_a==0:
            a_ruins+=1
        else:
            b_ruins+=1
    
    p_of_ruin = a_ruins/N

    return p_of_ruin
```

```python
probs=range(4,7)
a_vals=range(10,100,10)
labels=[str(x) for x in a_vals]
z=100
rounds=1000
ticks = range(len(a_vals))
fig,ax =  plt.subplots(len(probs),figsize=(30,30))
#fig.tight_layout(pad=3.0)

for i in probs:
    p=i/10
    q=1-p
    probs_of_ruin=[]
    probs_of_ruin_sim=[]
    capitals=[]
    for j in range(len(a_vals)):
        a=a_vals[j]
        b=z-a

        capitals.append(a)
        probs_of_ruin_sim.append(simulate_game(a,b,p,rounds))
        if p==0.5:
            probs_of_ruin.append(1-a/z)
        else:
            probs_of_ruin.append(((q/p)**a - (q/p)**z)/(1-(q/p)**z))
    
    ax[(i-4)].set_title("pA="+str(p))
    ax[(i-4)].bar(,probs_of_ruin, color = "black", width=0.2)
    ax[(i-4)+0.2].bar(,probs_of_ruin_sim, color = "blue", width=0.2)

plt.figure()
plt.show()
```
