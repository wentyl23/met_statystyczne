```python
import random
import matplotlib.pyplot as plt
import math
```

```python
def simulate_game(a,b,p,N,r):
    q=1-p
    balances=[]
    for i in range(N):
        rounds = 0
        curr_a = a
        curr_b = b
        while curr_a > 0 and curr_b > 0 and rounds <= r : 
            if random.random() < p :
                curr_a+=1
                curr_b-=1
            else:
                curr_a-=1
                curr_b+=1
            rounds+=1
        balances.append(curr_a)
    
    return balances
```

```python
a=50
b=50
z=a+b
probs=[1/2,1/5,4/5]
rounds=[2,10,20,50,80,100]
games=5000
plots=len(probs)*len(rounds)

fig,ax =  plt.subplots(plots,figsize=(10,40))
fig.tight_layout(pad=3.0)

for i in range(len(probs)):
    p=probs[i]
    q=1-p
    for j in range(len(rounds)): 
        r=rounds[j]
        balances = simulate_game(a,b,p,games,r)
        ax[j*len(probs) + i].hist(balances,density=True)
        ax[j*len(probs) + i].set_title("Prob of balances after n=" + str(r) + " rounds and p(a)=" + str(p))
    
plt.show()
```
