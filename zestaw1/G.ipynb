```python
import random
import matplotlib.pyplot as plt
import math
```

```python
def simulate_game(a,b,p,N):
    q=1-p
    balances = []
    wins = []
    curr_a = a
    curr_b = b
    curr_wins = 0
    rounds = 0 
    while curr_a > 0 and curr_b > 0:
        if random.random() < p :
            curr_a+=1
            curr_b-=1
            curr_wins+=1
        else:
            curr_a-=1
            curr_b+=1
        balances.append(curr_a)
        wins.append(curr_wins)
        rounds+=1

    return balances,wins,rounds
```

```python
a=50
b=50
z=a+b
probs=[1/2,1/5,4/5]
games=5
plots = len(probs)*2


fig,ax =  plt.subplots(plots,figsize=(10,30))
fig.tight_layout(pad=3.0)

for i in range(len(probs)):
    p=probs[i]
    ax[i*2].set_title("wins to rounds trajectory for p=" + str(p))
    ax[i*2+1].set_title("balances to rounds trajectory for p=" + str(p))
    ax[i*2].set_xlabel('rounds')
    ax[i*2+1].set_xlabel('rounds')
    for j in range(games):
        balances, wins, rounds = simulate_game(a,b,p,games)
        ax[i*2].step(range(rounds), wins)
        ax[i*2+1].step(range(rounds), balances)

#    ax[i].hist(sim_games,density=True, bins=30)
#    ax[i].set_title("average rounds for p=" + str(p) + " - " + str(average_rounds))
    
plt.show()
```
